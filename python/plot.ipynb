```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stability_case_files = ['./lat-dir-rolling-mode.txt', './lat-dir-spiral-mode.txt', './lat-dir-dutch-roll.txt']
trim_case_files = ['./lat-dir-trim-c_u=0.00001.txt', './lat-dir-trim-c_u=0.1.txt', './lat-dir-trim-c_u=2.txt']

def plot_state_evolution(fn, num_of_vars):
    with open(fn, 'r') as f:
        header = ''
        for line in f.readlines():
            header = line.strip()
            break

    var_names = header.split()

    #data = np.loadtxt(fn, skiprows=1)[:-1]
    data = np.genfromtxt(fn, skip_header=1, skip_footer=1)
    data[:, 1] /= 279.1 #Normalize speed
    for i in range(1, num_of_vars):
        plt.plot(data[:, 0], data[:, i], label=var_names[i])
        i += 1

    plt.legend(loc='upper right')
    plt.xlabel('time')
    plt.title(fn)
    plt.show()
    
'''
plt.plot(data[:, 0], data[:, 1], '-b', label='1')
plt.plot(data[:, 0], data[:, 2], '-r', label='2')
plt.legend(loc='upper right')
plt.show()
'''
```

```python
for fn in stability_case_files:
    plot_state_evolution(fn, 5) #Ignore control
```

```python
for fn in trim_case_files:
    plot_state_evolution(fn, 7)
```
