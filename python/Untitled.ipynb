```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fn = './lat-dir-trim-c_u=0.5.txt'
with open(fn, 'r') as f:
    header = ''
    for line in f.readlines():
        header = line.strip()
        break

var_names = header.split()
print(var_names)

data = np.loadtxt(fn, skiprows=1)
#np.true_divide(data[:, 1], 279.1) #Normalize speed
data[:, 1] /= 279.1
for i in range(1, len(var_names)):
    plt.plot(data[:, 0], data[:, i], label=var_names[i])
    i += 1

plt.legend(loc='upper right')
plt.show()
'''
plt.plot(data[:, 0], data[:, 1], '-b', label='1')
plt.plot(data[:, 0], data[:, 2], '-r', label='2')
plt.legend(loc='upper right')
plt.show()
'''
```
